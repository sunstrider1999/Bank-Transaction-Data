```python
import pandas as pd
import numpy as np
df = pd.read_excel('bank.xlsx')
df.head()
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Выберем только количественные столбцы
num_cols = ['WITHDRAWAL AMT', 'DEPOSIT AMT', 'BALANCE AMT']
corr_data = df[num_cols]

# Вычислим корреляционную матрицу
corr_matrix = corr_data.corr()
print(corr_matrix)

# Визуализируем корреляционную матрицу
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='YlOrRd')
plt.title('Корреляционная матрица')
plt.show()

# Построим диаграммы рассеяния
plt.figure(figsize=(12, 12))
for i, col1 in enumerate(num_cols):
    for j, col2 in enumerate(num_cols):
        plt.subplot(len(num_cols), len(num_cols), i*len(num_cols) + j + 1)
        plt.scatter(df[col1], df[col2])
        plt.xlabel(col1)
        plt.ylabel(col2)
plt.tight_layout()
plt.show()
```

```python
df = df.rename(columns={'Account No': 'Номер счета',
                       'DATE': 'Дата',
                       'TRANSACTION DETAILS': 'Детали транзакции',
                       'CHQ.NO.': 'Номер чека',
                       'VALUE DATE': 'Дата валютирования',
                       'WITHDRAWAL AMT': 'Сумма вывода средств',
                       'DEPOSIT AMT': 'Сумма депозита',
                       'BALANCE AMT': 'Сумма баланса',
                       '.': 'Дополнительная информация'})
```

```python
df.head()
```

```python
df['Дополнительная информация']
```

```python
df['Дополнительная информация'].value_counts()
```

```python
df =df.drop(['Дополнительная информация','Номер чека'],axis = 1)
```

```python
df.describe()
```

```python
df['День недели'] = df['Дата'].dt.day_name()
df.head()
```

```python
df['Количество операций в день'] = df.groupby(['Дата', 'Номер счета'])['Детали транзакции'].transform('count')
df.head()
```

```python
print(df.isnull().sum())
```

```python
most_common_value = df['Детали транзакции'].mode().iloc[0]
df['Детали транзакции'] = df['Детали транзакции'].fillna(most_common_value)
print(df.isnull().sum())
df.head()
```

```python
df['Разница между транзакцией и днем валютирования в днях'] = (df['Дата'] - df['Дата валютирования']).dt.days
df['Разница между транзакцией и днем валютирования в днях'].value_counts()
```

```python
df.head()
```

```python
if df['Сумма вывода средств'].isnull().any():
    mean_withdrawal = df['Сумма вывода средств'].mean()
    df['Сумма вывода средств'] = df['Сумма вывода средств'].fillna(mean_withdrawal)
if df['Сумма депозита'].isnull().any():
    mean_withdrawal = df['Сумма депозита'].mean()
    df['Сумма депозита'] = df['Сумма депозита'].fillna(mean_withdrawal)
df.head()
```

```python
print(df.isnull().sum())
```

```python
df['Детали транзакции'].value_counts()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

df['Детали транзакции'] = df['Детали транзакции'].astype(str)
tfidf = TfidfVectorizer(max_features=5000)
X = tfidf.fit_transform(df['Детали транзакции'])
tfidf_df = pd.DataFrame.sparse.from_spmatrix(X, columns=tfidf.get_feature_names_out())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
import pandas as pd

df['Детали транзакции'] = df['Детали транзакции'].astype(str)

tfidf = TfidfVectorizer(max_features=5000)
X = tfidf.fit_transform(df['Детали транзакции'])

tfidf_df = pd.DataFrame.sparse.from_spmatrix(X, columns=tfidf.get_feature_names_out())

svd = TruncatedSVD(n_components=10)
X_svd = svd.fit_transform(X)

svd_df = pd.DataFrame(X_svd, columns=[f'SVD_{i+1}' for i in range(10)])

df_with_svd = pd.concat([df, svd_df], axis=1)
df_with_svd.head()
```

```python
df = df_with_svd.drop(['Номер счета','Дата','Дата валютирования','Детали транзакции'], axis=1)
```

```python
df.head()
```

```python
days_of_week = ['Понедельник', 'Вторник', 'Среда', 'Четверг', 'Пятница', 'Суббота', 'Воскресенье']

# Создаем one-hot encoded матрицу с 0/1 значениями
one_hot_encoded = pd.get_dummies(df['День недели'], prefix='День_недели', columns=days_of_week, dtype=int)

# Объединяем one-hot encoded матрицу с исходным DataFrame
df = pd.concat([df, one_hot_encoded], axis=1)

# Удаляем исходный столбец "День недели"
df = df.drop(['День недели','День_недели_Friday'], axis=1)
```

```python
df.head()
```

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Выделение столбцов для стандартизации
cols_to_scale = ['Сумма вывода средств', 'Сумма депозита', 'Сумма баланса', 'Количество операций в день', 'Разница между транзакцией и днем валютирования в днях']

# Создание объекта StandardScaler
scaler = StandardScaler()

# Стандартизация столбцов
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])

# Просмотр обновленного DataFrame
df.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.model_selection import GridSearchCV


# Определяем диапазон значений для гиперпараметра n_clusters
param_grid = {'n_clusters': [2]}

# Создаем объект GridSearchCV для подбора гиперпараметров
kmeans = KMeans()
grid_search = GridSearchCV(kmeans, param_grid, cv=5)

# Обучаем модель и получаем лучшие гиперпараметры
grid_search.fit(df)
best_n_clusters = grid_search.best_params_['n_clusters']
print(f"Лучшее количество кластеров: {best_n_clusters}")

# Обучаем модель с лучшими гиперпараметрами
final_model = KMeans(n_clusters=best_n_clusters)
final_model.fit(X)

# Добавляем метки кластеров в DataFrame
df['Cluster'] = final_model.labels_

# Назовем кластеры
cluster_names = {0: 'Нормальная транзакция', 1: 'Мошенничество'}
df['Cluster_Name'] = df['Cluster'].map(cluster_names)

# Просматриваем результаты кластеризации
print(df.groupby('Cluster_Name').size())
```
