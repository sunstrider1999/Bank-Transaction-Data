```python
import matplotlib.pyplot as plt
import seaborn as sns

#    
num_cols = ['WITHDRAWAL AMT', 'DEPOSIT AMT', 'BALANCE AMT']
corr_data = df[num_cols]

#   
corr_matrix = corr_data.corr()
print(corr_matrix)

#   
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='YlOrRd')
plt.title(' ')
plt.show()

#   
plt.figure(figsize=(12, 12))
for i, col1 in enumerate(num_cols):
    for j, col2 in enumerate(num_cols):
        plt.subplot(len(num_cols), len(num_cols), i*len(num_cols) + j + 1)
        plt.scatter(df[col1], df[col2])
        plt.xlabel(col1)
        plt.ylabel(col2)
plt.tight_layout()
plt.show()
```

```python
df = df.rename(columns={'Account No': ' ',
                       'DATE': '',
                       'TRANSACTION DETAILS': ' ',
                       'CHQ.NO.': ' ',
                       'VALUE DATE': ' ',
                       'WITHDRAWAL AMT': '  ',
                       'DEPOSIT AMT': ' ',
                       'BALANCE AMT': ' ',
                       '.': ' '})
```

```python
df.head()
```

```python
df[' ']
```

```python
df[' '].value_counts()
```

```python
df =df.drop([' ',' '],axis = 1)
```

```python
df.describe()
```

```python
df[' '] = df[''].dt.day_name()
df.head()
```

```python
df['   '] = df.groupby(['', ' '])[' '].transform('count')
df.head()
```

```python
print(df.isnull().sum())
```

```python
most_common_value = df[' '].mode().iloc[0]
df[' '] = df[' '].fillna(most_common_value)
print(df.isnull().sum())
df.head()
```

```python
df['       '] = (df[''] - df[' ']).dt.days
df['       '].value_counts()
```

```python
df.head()
```

```python
if df['  '].isnull().any():
    mean_withdrawal = df['  '].mean()
    df['  '] = df['  '].fillna(mean_withdrawal)
if df[' '].isnull().any():
    mean_withdrawal = df[' '].mean()
    df[' '] = df[' '].fillna(mean_withdrawal)
df.head()
```

```python
print(df.isnull().sum())
```

```python
df[' '].value_counts()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

df[' '] = df[' '].astype(str)
tfidf = TfidfVectorizer(max_features=5000)
X = tfidf.fit_transform(df[' '])
tfidf_df = pd.DataFrame.sparse.from_spmatrix(X, columns=tfidf.get_feature_names_out())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
import pandas as pd

df[' '] = df[' '].astype(str)

tfidf = TfidfVectorizer(max_features=5000)
X = tfidf.fit_transform(df[' '])

tfidf_df = pd.DataFrame.sparse.from_spmatrix(X, columns=tfidf.get_feature_names_out())

svd = TruncatedSVD(n_components=10)
X_svd = svd.fit_transform(X)

svd_df = pd.DataFrame(X_svd, columns=[f'SVD_{i+1}' for i in range(10)])

df_with_svd = pd.concat([df, svd_df], axis=1)
df_with_svd.head()
```

```python
df = df_with_svd.drop([' ','',' ',' '], axis=1)
```

```python
df.head()
```

```python
days_of_week = ['', '', '', '', '', '', '']

#  one-hot encoded   0/1 
one_hot_encoded = pd.get_dummies(df[' '], prefix='_', columns=days_of_week, dtype=int)

#  one-hot encoded    DataFrame
df = pd.concat([df, one_hot_encoded], axis=1)

#    " "
df = df.drop([' ','__Friday'], axis=1)
```

```python
df.head()
```

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

#    
cols_to_scale = ['  ', ' ', ' ', '   ', '       ']

#   StandardScaler
scaler = StandardScaler()

#  
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])

#   DataFrame
df.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.model_selection import GridSearchCV


#      n_clusters
param_grid = {'n_clusters': [2]}

#   GridSearchCV   
kmeans = KMeans()
grid_search = GridSearchCV(kmeans, param_grid, cv=5)

#      
grid_search.fit(df)
best_n_clusters = grid_search.best_params_['n_clusters']
print(f"  : {best_n_clusters}")

#     
final_model = KMeans(n_clusters=best_n_clusters)
final_model.fit(X)

#     DataFrame
df['Cluster'] = final_model.labels_

#  
cluster_names = {0: ' ', 1: ''}
df['Cluster_Name'] = df['Cluster'].map(cluster_names)

#   
print(df.groupby('Cluster_Name').size())
```
